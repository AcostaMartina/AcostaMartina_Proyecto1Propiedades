```python
import json
from bs4 import BeautifulSoup
import requests

result = []
tipos = ["casas","chacras","apartamentos"]
for tipo in tipos:
    url = f"https://www.sigaloavarela.com/{tipo}/en-venta"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    propiedades = soup.find_all("div", class_="col-xl-4")

    for propiedad in propiedades:
        precio = propiedad.find_all ("a", class_= "sale_btn")
        descripcion = propiedad.find ("p")
        result.append({"type": tipo, "price":precio[1].text, "description":descripcion.text})

        print(precio[1].text, descripcion.text)



```

```python
with open ("propiedades.json", "w") as out_file:
    json.dump(result, out_file, indent=4)
```
